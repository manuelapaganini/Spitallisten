# Bern Datahandling

Vorgängig habe ich das PDF mit Acrobat Reader zu einem Excel zurückverwandelt, darin die grün ausgefüllten Felder mit Text markiert, der Tabelle einen eindeutigen Header gegeben und das Ganze als CSV exportiert. 

Hier werden die verschiedenen Werte zu 0 und 1 umgewandelt.
1. NaN-Values
2. True, hängige Beschwerde, IVHSM-Leistungsauftrag, Teil-LA, Dialyse
3. alles was 'befristet' enthält

```python
# Import stuff
import pandas as pd
import re
```

### PDF-Import

```python
df = pd.read_csv('Listen/Bern_6.3.csv')
```

```python
df.head(3)
```

### Datenbereinigung: Werte zu 0 und 1 umwandeln

```python
# NaNs zu 0 umwandeln
df=df.fillna(0)
df.head(3)
```

```python
#Text zu 0 und 1 umwandeln

cleanup_dict = {'TRUE':1, True:1, 'Dialyse1':1, 'IVHSM-Leistungsauftrag\t\t':1, 'Teil-LA2':1, 'hängige Beschwerde':0}

df.replace(cleanup_dict, inplace=True)
df.head(3)
```

```python
# Regex-Funktion für das Umwandlen von befristet

def search_befristet(cell):
    if type(cell) == str:
        if re.match(r'befristet', cell) or re.match(r'Teil-LA2', cell):
            return 1
        else:
            return cell
    else:
        return cell
```

```python
# Funktion auf das DF anwenden

columns = df.columns.tolist()[4:]
for column in columns:
    df[column] = df[column].apply(search_befristet)
#df
```
